```python
import string
from string import punctuation
import re, os
from gensim.summarization.textcleaner import split_sentences
from pymystem3 import Mystem
mystem = Mystem()
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
```

```python
os.getcwd()
```

```python
with open ('city_smells.txt', "r", encoding = "utf-8") as text:
    text = text.read()
print(text)
```

```python
# Удалим часть текста, которая не является непосредственно текстом:
new_text = text.replace("Источники:\nСтраница проекта Smelly Maps\nИнтерактивные карты проекта\nНелли Бурцева", "")
print(new_text)
```

```python
# Разобьем текст на предложения:
text_sent = list(split_sentences(new_text))
text_sent
```

```python
# Посчитаем среднюю длину предложения в тексте:
average_sent = sum(len(sentence) for sentence in text_sent) / len(text_sent)
print ("Cредняя длина предложения в тексте:", average_sent)
```

```python
# Посчитаем, во сколько раз самое длинное предложение в тексте длиннее самого короткого:
average_max_min_sent = max(len(sentence) for sentence in text_sent) / min(len(sentence) for sentence in text_sent)
print("Cамое длинное предложение длиннее самого короткого в", average_max_min_sent, "раз")
```

```python
string.punctuation = string.punctuation + '«»—…“”'
```

```python
# Посчитаем среднее количество пунктуационных знаков в предложении:
text_punct = [[punct for punct in nested if punct in string.punctuation] for nested in text_sent]
average_punct = sum(len(punct) for punct in text_punct) / len(text_sent)
print("Cреднее количество пунктуационных знаков в предложении:", average_punct)
```

```python
# Перейдем к подсчету слов (при этом будем считать только слова, написанные на русском языке)
# Перед подсчетом предобработаем текст (удалим пунктуацию, цифры и английские слова) и разделим на токены:
def preprocess_tokenize(new_text):
    text_preprocessed_tokenized = []
    for sentence in new_text:
        clean_words = [word.strip(string.punctuation) for word in word_tokenize(new_text)]
        clean_words = [word for word in clean_words if word]
#         Разделим слова, соединенные знаками препинания('выбросы/выхлопы', 'слова-метки') на 2 слова (чтобы они не посчитались как 1 слово)    
        clean_words = [re.sub('(?<=[а-я])([-/])(?=[а-я])', r' \1 ', i) for i in clean_words]
        clean_words = [word for wordset in clean_words for word in wordset.split(' ')]
        r = re.compile("[а-яА-Я]")
        clean_words = [word for word in filter(r.match, clean_words)]
        clean_words = [word.lower() for word in clean_words if word]
        return clean_words
words = preprocess_tokenize(new_text)
words
```

```python
# Посчитаем среднюю длину русского слова в тексте:
average_word = sum(len(word) for word in words) / len(words)
print ("Cредняя длина русского слова в тексте:", average_word)
```

```python
# Посчитаем, во сколько раз самое длинное русское слово длиннее самого короткого:
average_max_min_word = max(len(word) for word in words) / min(len(word) for word in words)
print ("Cамое длинное русское слово длиннее самого короткого в", average_max_min_word, "раз")
```

```python
# Разделим текст на леммы и сделаем из списка множество с неповторяющимися русскими словами:
words_analized = mystem.analyze(new_text)
words_analized_l = [parse['analysis'][0]['lex'] for parse in words_analized if parse.get('analysis')]
text_lem = set(words_analized_l)
text_lem
```

```python
# Подсчитаем количество уникальных (неповторяющихся) русских слов в тексте:
print("Количество уникальных русских слов в тексте:", len(text_lem))
```

```python
# Подсчитаем отношение количества уникальных русских слов к общему количеству русских слов в тексте:
average_unique_word = len(text_lem) / len(words)
print("Отношение количества уникальных русских слов к общему количеству русских слов в тексте:", average_unique_word)
```

```python
# Выведем список стоп-слов:
stops = stopwords.words('russian')
print(stops)
```

```python
# Добавим в него еще несколько стоп-слов:
stops.append('который')
stops.append('например')
stops.append('также')
print(stops)
```

```python
# Удалим из списка слов стоп-слова:
words_analized_2 = [word for word in words_analized_l if word not in stops]
words_analized_2
```

```python
# Выведем список самых частотных русских слов:
from operator import itemgetter
sorted_frequency_table = sorted(FreqDist(words_analized_2).items(), key=itemgetter(1), reverse=True)    
for word, freq in sorted_frequency_table[:10]:
    print(' '.join((word, str(freq))))
```

```python
# Найдем самое частотное русское слово в тексте (без соотношения к общему количеству слов):
print("Самое частотное русское слово в тексте:", sorted_frequency_table[0][0], "(встречается", sorted_frequency_table[0][1], "раз)")
```
